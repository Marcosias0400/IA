```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Condiciones iniciales
alpha=1/10**4
beta=1/10
# Tolerancia
epsilon=1/10**3
# maxima cantidad de iteraciones
N=50
# Punto inicial
x_0=np.array([3,-2])
```

```python
# funcion de prueba
def f(x:tuple,C:np.float64)->np.float64:
  return np.float64(2*x[0]**2+2*x[0]*x[1]+3*x[1]**2/2-x[0]+x[1]+C)
```

```python
def d1f(x:np.array)->np.array:
  return np.array([4*x[0]+2*x[1]-1,3*x[1]+2*x[0]+1])
```

```python
def norm2A(g:np.array,A:np.array)->np.float64:
  return np.float64(np.dot(np.dot(g,A),g))
```

```python
def des_grad(alpha:np.float64,beta:np.float64,tol:np.float64,MAX_IT:np.int32,x_0:np.array)->list:
  A=np.array([[4,2],[2,3]])
  k=0
  x=x_0
  e=0
  d=d1f(x)
  reg=[k,x,d,f(x,1),1,e]
  c=1
  while np.linalg.norm(d1f(x))>=tol and k<=MAX_IT:
    d=d1f(x)
    xant=x
    lamb=np.dot(d,d)/norm2A(d,A)
    x=x+lamb*d
    e=f(x,c)-f(xant,c)
    k+=1
    reg.append([k,x,d,f(x,c),lamb,e])
  return reg
l=des_grad(alpha,beta,epsilon,N,x_0)
```

```python
for o in l:
  print(o)
```
